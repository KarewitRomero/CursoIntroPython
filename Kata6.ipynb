## Ejercicio 1

En primer lugar, crea una variable denominada `planets`. Agrega los ocho planetas (sin Plutón) a la lista. A continuación, muestra el número de planetas.

```python
# Creamos la lista planets y la mostramos
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Nepturno']
print("Son", len(planets), "planetas")
```

Agregar a Plutón

```python
# Agregamos a plutón y mostramos el último elemento
planets.append('Plutón')
print('El último planeta es:', planets[-1])
print("Son", len(planets), "planetas")
```

## Ejercicio 2

```python
# Lista de planetas
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Nepturno']
```

Solicitar información al usuario

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
seleccionPlaneta = input("Introduce el nombre del planeta")
```

Encontrar planeta en la lista

```python
# Busca el planeta en la lista
planeta_indice = planetas.index(seleccionPlaneta)
```

Mostrar planetas cercanos de acuerdo con la entrada del usuario

```python
# Muestra los planetas más cercanos al sol
print('Los planetas más cercanos al sol que', seleccionPlaneta, 'son', planetas[0:planeta_indice])
```

Planetas más alejados

```python
# Muestra los planetas más lejanos al sol
print('Los planetas más lejanos al sol que', seleccionPlaneta, 'son', planetas[planeta_indice + 1:])
```
