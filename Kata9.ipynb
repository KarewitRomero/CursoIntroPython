## Ejercicio 1 
Trabajar con argumentos en funciones

```python
#Función para leer 3 tanques de combustible y muestre el promedio
def informe(tanque_fijo, tanque_movible, tanque_pipa):
    promedio = (tanque_fijo + tanque_movible + tanque_pipa) / 3
    return  f"""Informe de combustible:
    Tanque fijo: {tanque_fijo}%
    Tanque movible: {tanque_movible}%
    Tanque de pipa: {tanque_pipa}% 
    Promedio total: {promedio}%
    """
```

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(informe(90, 100, 70))
```

```python
# Función promedio 
def prom_combustible(values):
    total = sum(values)
    numero_tanques = len(values)
    return total / numero_tanques

prom_combustible([90, 100, 70]) 
```

```python
# Actualiza la función
def informe(tanque_fijo, tanque_movible, tanque_pipa):
    return  f"""Informe de combustible:
    Tanque fijo: {tanque_fijo}%
    Tanque movible: {tanque_movible}%
    Tanque de pipa: {tanque_pipa}% 
    Promedio total: {prom_combustible([tanque_fijo, tanque_movible, tanque_pipa])}%
    """

print(informe(70, 90, 100))
```

Trabajar con argumentos de palabras clave en funciones

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno
def informeMision(prelanzamienzo, tiempo_vuelo, destino, tanque_externo, tanque_interno):
    return f"""
    Destino de la misión: {destino}
    Tiempo de viaje: {prelanzamienzo + tiempo_vuelo} min.
    Total de combustible: {tanque_externo + tanque_interno} lts.
    """

print(informeMision(5, 120, "Saturno", 12000000, 3000000))
```

```python
# Escribe tu nueva función de reporte considerando lo anterior
def informeMision(destino, *minutos, **deposito_combustible):
    return f"""
    Destino: {destino}
    Tiempo de viaje: {sum(minutos)} min.
    Total de combustible: {sum(deposito_combustible.values())} lts
    """

print(informeMision("Jupiter", 5, 43, externo = 1500000, interno=23000000))
```

```python
# Escribe tu nueva función
def informeMision(destino, *minutos, **deposito_combustible):
    InformeFinal = f"""
    Destino: {destino}
    Tiempo de viaje: {sum(minutos)} min.
    Total de combustible: {sum(deposito_combustible.values())} lts
    """
    for tanque, litros in deposito_combustible.items():
        InformeFinal += f"El tanque {tanque} contiene {litros} litros   -----     "
    return InformeFinal

print(informeMision("Venus", 8, 59, externo = 1500000, interno=23000000))
```
