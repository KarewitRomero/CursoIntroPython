## Ejercicio 1

Exploremos cómo podemos crear un programa que pueda calcular la distancia entre dos planetas.  Comenzaremos usando dos distancias de planetas: Tierra (149.597.870 km) y Júpiter (778.547.200 km).

```python
# Crear variables para almacenar las dos distancias
planeta_Tierra = 149597870
planeta_Jupiter = 778547200

```

Operación

A continuación, puedes convertir la distancia del kilómetro en millas. 1km = 0.621371 millas

```python
# Calcular la distancia entre planetas
distanciaEntrePlanetas = (planeta_Jupiter - planeta_Tierra) * 0.621371
print("La distancia que hay entre el planeta Tierra y Júpiter es de: " + str(distanciaEntrePlanetas) + " millas")
```

## Ejercicio 2

```python
# Almacenar las entradas del usuario
primerPlaneta = input("Ingresa la distancia del primer planeta (en referencia al sol): ")
segundoPlaneta = input('Ingresa la distancia del segundo planeta (en referencia al sol): ')
#Pista: variable = input("¿Cuál es tu nombre?")
```

Conversión de datos

```python
# Convierte las cadenas de ambos planetas a números enteros
primerPlaneta = int(primerPlaneta)
segundoPlaneta = int(segundoPlaneta)
```

Con los valores almacenados como números, ahora puedes agregar el código para realizar el cálculo, restando el primer planeta del segundo. Debido a que el segundo planeta podría ser un número mayor, usarás `abs` para convertirlo a un valor absoluto. También agregarás el código para mostrar el resultado en millas multiplicando la distancia del kilómetro por 0.621371

```python
# Realizar el cálculo y determinar el valor absoluto
distanciaPlanetaria = segundoPlaneta - primerPlaneta
print(abs(distanciaPlanetaria))

# Convertir de KM a Millas
km_millas = distanciaPlanetaria * 0.621371
print(abs(km_millas))
```
