## Ejercicio 1

```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""
```

Dividir texto

```python
# Añade el código necesario
dividirTexto = text.split('. ')
dividirTexto
```

Palabras clave

```python
# Define las palabras pista: average, temperature y distance suenan bien
palabras_clave = ["average", "temperature", "distance"]
```

Bucle

```python
# Ciclo for para recorrer la cadena
for datos in dividirTexto:
    for palabra_clave in palabras_clave:
        if palabra_clave in datos:
            print(datos)
            break
```

Bucle para reemplazar palabras

```python
# Ciclo para cambiar C a Celsius
for datos in dividirTexto:
    for palabra_clave in palabras_clave:
        if palabra_clave in datos:
            print(datos.replace(' C', ' Celsius'))
            break
```

## Ejercicio 2

```python
# Datos con los que vas a trabajar
name = "Moon"
gravity = 0.00162 # in kms
planet = "Earth"
```

Crear titulo

```python
# Creamos el título
titulo = f'Gravity of the {name} and {planet}'
print(titulo)
```

Plantilla

```python
# Creamos la plantilla
plantilla = f""" 
Name of the planet: {planet} 
The gravity on the {name} is: {gravity * 1000} m/s^2 
"""
print(plantilla)
```

Unir titulo y plantilla

```python
union_cadenas = f"""{titulo.title()} 
{plantilla} 
""" 
print(union_cadenas)
```

Nuevos datos (Comprobación de plantilla)

```python
# Comprueba la plantilla
name = "Ganímedes"
gravity = 0.00143
planet = "Marte"

# print(nombre_plantilla)
print(plantilla)
```

Nueva plantilla

```python
# Nueva plantilla
nuevaPlantilla = """
Nombre del planeta: {planet}
La gravedad en {name} es de: {gravity} m/s^2
"""
print(nuevaPlantilla.format(name=name, planet=planet, gravity=gravity))
```

Corrección de la gravedad

```python
# Pista: print(nueva_plantilla.format(variables))
print(nuevaPlantilla.format(name=name, planet=planet, gravity=gravity*1000))
```
