Crear diccionario que almacene información sobre el planeta Marte

```python
#Crea un diccionario llamado planet con los datos propuestos
planet = {
    'name': 'Mars',
    'moons': 2
}
```

```python
#Muestra el nombre del planeta y el número de lunas que tiene
print(f'{planet["name"]} tiene {planet["moons"]} lunas')
```

```python
#Agrega la clave circunferencia con los datos proporcionados previamente
planet['circunferencia (km)'] = {
    'polar': 6752,
    'equatorial': 6792
}
```

```python
#Imprime el nombre del planeta con su circunferencia polar
print(f'{planet["name"]} tiene una circunferencia polar de {planet["circunferencia (km)"]["polar"]} km')
```

## Ejercicio 2

Cálculo de valores

```python
# Planets and moons

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
#Añade el código para determinar el número de lunas
lunas = planet_moons.values()
planetas = len(planet_moons.keys())
```

```python
#Agrega el código para contar el número de lunas
lunas_totales = 0
for luna in lunas:
    lunas_totales = lunas_totales + luna

promedio = lunas_totales / planetas
print(lunas_totales)
print(promedio)
```
